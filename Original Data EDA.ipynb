```python
# prompt: Water Binder Ratio	OPC	PPC	DAYS	CO2	STRENGTH are the columns in dataframe generate explortery data analysis

import pandas as pd
import matplotlib.pyplot as plt


import seaborn as sns

df = pd.read_csv('Data.csv')
print(df.head())  # Display the first few rows of the DataFrame
df.tail()

# Data types and missing values
print(df.info())
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())

# Visualizations
# Histograms
df.hist(figsize=(10, 8))
plt.show()
```

```python
import matplotlib
matplotlib.use('Agg')
%matplotlib inline

import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'DeJavu Serif'
plt.rcParams['font.serif'] = ['Helvetica']
plt.rcParams.update({'font.size': 15})

# Define a professional style
plt.style.use('seaborn-muted')

# Calculate layout dimensions
num_columns = len(df.columns)
rows = (num_columns + 2) // 3  # Adjust for a grid of up to 3 columns per row

# Create the box plots
fig, axes = plt.subplots(nrows=rows, ncols=3, figsize=(12, rows * 4))
axes = axes.flatten()  # Flatten for easier indexing

for i, column in enumerate(df.columns):
    df.boxplot(column=column, ax=axes[i], notch=True, patch_artist=True)
    axes[i].set_title(column, fontsize=12)

# Hide unused axes
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

# Add a watermark in the center
plt.figtext(0.5, 0.5, 'Ashiq Nazir Bhat', ha='center', va='center', fontsize=12, alpha=0.2, color='gray', rotation=30)

# Save the plot as a professional-looking PDF with 1200 DPI
plt.savefig('Dataplots.pdf', format='pdf', dpi=1200)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a professional-looking heatmap with gridlines
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    cbar_kws={'shrink': 0.8},
    square=True,
    linewidths=0.5,  # Adds gridlines
    linecolor='gray'  # Sets the gridline color
)

# Add a watermark in the center
plt.figtext(0.5, 0.5, 'Ashiq Nazir Bhat', ha='center', va='center', fontsize=12, alpha=0.2, color='gray', rotation=30)

# Save the plot as a PDF with 1200 DPI
plt.tight_layout()
plt.savefig('correlationmatrix.pdf', format='pdf', dpi=1200)

# Show the plot
plt.show()
```
