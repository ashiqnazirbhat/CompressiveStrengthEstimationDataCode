```python
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from scipy.spatial.distance import euclidean
import pandas as pd
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def create_generator(noise_dim, output_dim):
    """
    Create generator neural network
    """
    return nn.Sequential(
        nn.Linear(noise_dim, 128),
        nn.ReLU(),
        nn.Linear(128, 256),
        nn.ReLU(),
        nn.Linear(256, output_dim)
    )

def create_critic(input_dim):
    """
    Create critic neural network
    """
    return nn.Sequential(
        nn.Linear(input_dim, 256),
        nn.ReLU(),
        nn.Linear(256, 128),
        nn.ReLU(),
        nn.Linear(128, 1)
    )

def train_critic(
    critic, 
    generator, 
    real_data, 
    noise_dim, 
    batch_size, 
    c_optimizer
):
    """
    Train the critic for one iteration
    """
    # Real data
    real_indices = torch.randperm(real_data.size(0))[:batch_size]
    real_samples = real_data[real_indices]
    
    # Generate fake data
    noise = torch.randn(batch_size, noise_dim)
    fake_samples = generator(noise).detach()
    
    # Critic forward pass
    critic_real = critic(real_samples)
    critic_fake = critic(fake_samples)
    
    # Wasserstein loss for critic
    critic_loss = torch.mean(critic_fake) - torch.mean(critic_real)
    
    # Optimize critic
    c_optimizer.zero_grad()
    critic_loss.backward()
    c_optimizer.step()
    
    # Weight clipping
    for p in critic.parameters():
        p.data.clamp_(-0.01, 0.01)
    
    return critic_loss.item()

def train_generator(
    critic, 
    generator, 
    noise_dim, 
    batch_size, 
    g_optimizer
):
    """
    Train the generator for one iteration
    """
    # Generate fake samples
    noise = torch.randn(batch_size, noise_dim)
    fake_samples = generator(noise)
    
    # Generator loss
    g_loss = -torch.mean(critic(fake_samples))
    
    # Optimize generator
    g_optimizer.zero_grad()
    g_loss.backward()
    g_optimizer.step()
    
    return g_loss.item()
```

```python
def train_wgan(
    num_epochs=1000, 
    batch_size=64, 
    noise_dim=1, 
    critic_iterations=5
):
    """
    Main training function for Wasserstein GAN
    """
    # Prepare data
    real_data = generate_real_data(10000)
    
    # Initialize models
    generator = create_generator(noise_dim, 1)
    critic = create_critic(1)
    
    # Optimizers (using RMSprop as suggested in original WGAN paper)
    g_optimizer = optim.RMSprop(generator.parameters(), lr=0.00005)
    c_optimizer = optim.RMSprop(critic.parameters(), lr=0.00005)
    
    # Training logs
    g_losses = []
    c_losses = []
    
    # Training loop
    for epoch in range(num_epochs):
        # Train Critic multiple times
        epoch_c_losses = []
        for _ in range(critic_iterations):
            critic_loss = train_critic(
                critic, generator, real_data, 
                noise_dim, batch_size, c_optimizer
            )
            epoch_c_losses.append(critic_loss)
        
        # Train Generator
        generator_loss = train_generator(
            critic, generator, noise_dim, 
            batch_size, g_optimizer
        )
        
        # Store losses
        c_losses.append(np.mean(epoch_c_losses))
        g_losses.append(generator_loss)
        
        # Print progress
        if epoch % 100 == 0:
            print(f'Epoch [{epoch}/{num_epochs}], '
                  f'Critic Loss: {np.mean(epoch_c_losses):.4f}, '
                  f'Generator Loss: {generator_loss:.4f}')
    
    return generator
```

```python
def generate_real_data(num_samples):
    """
    Generate real data from underlying regression function
    """
    x = np.random.uniform(0, 10, num_samples)
    y = 2 * x + 1 + np.random.normal(0, 1, num_samples)
    return torch.FloatTensor(y.reshape(-1, 1))
```

```python
def generate_synthetic_data(generator, num_samples=10000):
    """
    Generate synthetic data using trained generator
    """
    noise = torch.randn(num_samples, 1)
    synthetic_data = generator(noise).detach().numpy().flatten()
    return synthetic_data
```

```python
def compute_descriptive_statistics(real_data, synthetic_data):
    """
    Compare basic descriptive statistics
    """
    print("\n--- Descriptive Statistics Comparison ---")
    print("Real Data:")
    print(f"Mean: {np.mean(real_data):.4f}")
    print(f"Median: {np.median(real_data):.4f}")
    print(f"Standard Deviation: {np.std(real_data):.4f}")
    print(f"Skewness: {stats.skew(real_data):.4f}")
    print(f"Kurtosis: {stats.kurtosis(real_data):.4f}")
    
    print("\nSynthetic Data:")
    print(f"Mean: {np.mean(synthetic_data):.4f}")
    print(f"Median: {np.median(synthetic_data):.4f}")
    print(f"Standard Deviation: {np.std(synthetic_data):.4f}")
    print(f"Skewness: {stats.skew(synthetic_data):.4f}")
    print(f"Kurtosis: {stats.kurtosis(synthetic_data):.4f}")

def perform_statistical_tests(real_data, synthetic_data):
    """
    Perform statistical tests to compare distributions
    """
    print("\n--- Statistical Tests ---")
    
    # Kolmogorov-Smirnov Test
    ks_statistic, ks_pvalue = stats.ks_2samp(real_data, synthetic_data)
    print("Kolmogorov-Smirnov Test:")
    print(f"Statistic: {ks_statistic:.4f}")
    print(f"P-value: {ks_pvalue:.4f}")
    print("Interpretation: Low p-value suggests different distributions")

def visualize_data_comparison(real_data, synthetic_data):
    """
    Create comparative visualizations
    """
    plt.figure(figsize=(15, 5))
    
    # Histograms
    plt.subplot(131)
    plt.hist(real_data, bins=50, alpha=0.5, label='Real Data')
    plt.hist(synthetic_data, bins=50, alpha=0.5, label='Synthetic Data')
    plt.title('Distribution Comparison')
    plt.legend()
    
    # Q-Q Plot
    plt.subplot(132)
    stats.probplot(real_data, plot=plt)
    plt.title('Real Data Q-Q Plot')
    
    plt.subplot(133)
    stats.probplot(synthetic_data, plot=plt)
    plt.title('Synthetic Data Q-Q Plot')
    
    plt.tight_layout()
    plt.show()
```
