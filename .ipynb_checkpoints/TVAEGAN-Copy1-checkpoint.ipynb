# Generate New Data

```python
from sdv.metadata import Metadata

metadata = Metadata.detect_from_dataframe(
    data=df,
    table_name='data')
```

```python
metadata
```

```python
from sdv.single_table import TVAESynthesizer

# Initialize the TVAE synthesizer with the desired configuration
synthesizer = TVAESynthesizer(
    metadata=metadata,  # Metadata is required to describe the dataset
    epochs=10000,  # Increase epochs for better training on small datasets
    batch_size=64,  # Training batch sizer
    verbose=True  # Print training progress
)
```

```python
synthesizer.fit(df)
```

### Step 3: Generate synthetic data

```python
synthetic_data = synthesizer.sample(num_rows=10000)
```

```python
import shap
import matplotlib.pyplot as plt

# Assuming shap_values and X are already defined
# Generate the force plot for the first prediction
force_plot = shap.force_plot(
    shap_values.base_values[0],  # Base value of the model's output
    shap_values.values[0],       # SHAP values for the first row
    X.iloc[0, :]                 # Feature values for the first row
)

# Save the force plot as a PDF
# First, we use `shap.save_html` to save it as an HTML file
shap.save_html("force_plot.html", force_plot)

# Then, we can use a library like `pdfkit` to convert the HTML to PDF, or use Matplotlib to save it as an image.
# To directly save it as a PDF, we will use the html2pdf approach:

import pdfkit
pdfkit.from_file("force_plot.html", "force_plot.pdf")

# The force plot is now saved as 'force_plot.pdf'
```

```python
import shap
import matplotlib.pyplot as plt
import os

# Directory to save the plots
output_dir = "dependence_plots"
os.makedirs(output_dir, exist_ok=True)

# Loop through all features in the dataset
for feature in X.columns:
    print(f"Generating dependence plot for {feature}...")
    shap.dependence_plot(feature, shap_values.values, X)
    
    # Save each plot
    plt.savefig(f"{output_dir}/{feature}_dependence_plot.pdf", bbox_inches="tight")
    plt.show()

print(f"Dependence plots saved to: {output_dir}")
```
