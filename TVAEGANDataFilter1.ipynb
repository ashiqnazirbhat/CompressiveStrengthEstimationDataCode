```python
from sklearn.ensemble import RandomForestRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
# Step 1: Train a Baseline Model on Original Data
X_original = df1.drop(columns=['STRENGTH'])
y_original = df1['STRENGTH']
```

```python
model = LGBMRegressor(random_state=42)
model.fit(X_original, y_original)
```

```python
# Evaluate baseline model performance
original_predictions = model.predict(X_original)
baseline_r2 = r2_score(y_original, original_predictions)
baseline_mae = mean_absolute_error(y_original, original_predictions)
baseline_mse = mean_squared_error(y_original, original_predictions)
```

```python
# Step 2: Predict on Synthetic Data and Calculate Errors
X_synthetic = df2.drop(columns=['STRENGTH'])
y_synthetic = df2['STRENGTH']
```

```python
synthetic_predictions = model.predict(X_synthetic)
synthetic_errors = abs(y_synthetic - synthetic_predictions)
```

```python
# Step 3: Filter Synthetic Data Based on Prediction Errors
# Select samples with error less than or equal to the maximum error in original data
error_threshold = max(abs(y_original - original_predictions))
error_threshold = error_threshold-5
error_threshold
```

```python
filtered_synthetic_data = df2[synthetic_errors <= error_threshold]
```

```python
filtered_synthetic_data
```

```python
# Step 4: Combine Original and Filtered Synthetic Data
final_data = pd.concat([df1, filtered_synthetic_data], ignore_index=True)
```
